```python
# Merging in Pandas
import pandas as pd
```

```python
staff_df = pd.DataFrame([{'Name': 'Teja', 'Role': 'Developer'},
                         {'Name': 'Nag', 'Role': 'Analyst'},
                         {'Name': 'Naveen', 'Role': 'Admin'},
                         {'Name': 'bobby', 'Role': 'Analyst'}])
staff_df
```

```python
staff_df = staff_df.set_index('Name')
staff_df
```

```python
student_df = pd.DataFrame([{'Name': 'Naveen', 'School': 'Andhra University'},
                           {'Name': 'Nag', 'School': 'Gitam University'},
                           {'Name': 'Teja', 'School': 'JNTUK'}])
student_df
```

```python
student_df = student_df.set_index('Name')
student_df
```

```python
#outer join 
pd.merge(staff_df, student_df, how='outer', left_index=True, right_index=True)
```

```python
#inner join 
pd.merge(staff_df, student_df, how='inner', left_index=True, right_index=True)
```

```python
# left join 
pd.merge(staff_df, student_df, how='left', left_index=True, right_index=True)
```

```python
# right join 
pd.merge(staff_df, student_df, how='right', left_index=True, right_index=True)
```

```python
# resetting the index
staff_df = staff_df.reset_index('Name')
staff_df
```

```python
# restting the index
student_df = student_df.reset_index('Name')
student_df
```

```python
# joining by using column names 
pd.merge(staff_df, student_df, how='outer', left_on='Name', right_on='Name')
```

```python
staff_df = pd.DataFrame([{'Name': 'Teja','age': 22, 'Role': 'Developer'},
                         {'Name': 'Nag','age': 23, 'Role': 'Analyst'},
                         {'Name': 'Naveen','age': 24, 'Role': 'Admin'},
                         {'Name': 'bobby','age': 25, 'Role': 'Analyst'},
                         {'Name': 'Teja','age': 30, 'Role': 'Developer'}])
staff_df
```

```python
student_df = pd.DataFrame([{'Name': 'Naveen','age': 24, 'School': 'Andhra University'},
                           {'Name': 'Nag','age': 27, 'School': 'Gitam University'},
                           {'Name': 'Teja','age': 28, 'School': 'JNTUK'}])
student_df
```

```python
# joining by using two column names 
pd.merge(staff_df, student_df, how='inner', left_on=['Name','age'], right_on=['Name','age'])
```

```python
# Pivot Tables
import numpy as np
staff_df.pivot_table(values= 'age', index='Name',columns= 'Role',aggfunc=np.sum)
```

```python
#concatenating
df = pd.concat([staff_df,student_df])
df
```

```python
df = pd.concat([staff_df,student_df], axis = 1)
df
```
